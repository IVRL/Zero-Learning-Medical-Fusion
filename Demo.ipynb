```python
import numpy as np
from imageio import imread

import matplotlib.pyplot as plt
%matplotlib inline

import vggfusiongpu as fu

from torchvision.models.vgg import vgg19
model = vgg19(True).cuda().eval()
```

```python
# Example with grayscale data
# im1 = imread('images/MRI-CT/mr.png')
# im2 = imread('images/MRI-CT/ct.png')

# Example with color and grayscale
im1 = imread('images/MRI-SPECT/mr.png')
im2 = imread('images/MRI-SPECT/tc.png')

fusion = fu.fuse([im1, im2], model, False)
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
if im1.ndim == 3:
    plt.imshow(im1)
else:
    plt.imshow(im1, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
if im2.ndim == 3:
    plt.imshow(im2)
else:
    plt.imshow(im2, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
if fusion.ndim == 3:
    plt.imshow(fusion)
else:
    plt.imshow(fusion, cmap='gray')
plt.axis('off')
```
